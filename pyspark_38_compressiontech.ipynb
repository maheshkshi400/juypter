```python
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder.appName("https://github.com/maheshkshi400/MK").getOrCreate()


df = spark.read.format('csv') \
    .option('inferschema', True) \
    .option('header',True)\
    .option('sep',",")\
    .load('C:/Users/Pc/juypter/empdf12.csv')
df.show()
```

```python
# to write the data n compression by using snappy technique

df.write.format("parquet").option("compression","snappy").save("C:/Users/Pc/juypter/snappy-parquet")
```
