```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var= pd.read_csv('tips.csv')
var
```

```python
sns.catplot(x="size",y="tip",data=var,palette="terrain",edgecolor="r",kind="bar")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,palette="terrain_r",edgecolor="r",kind="swarm")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,palette="terrain_r",edgecolor="r",kind="bar",hue="sex")
plt.show()
```

```python
sns.catplot(x="tip",y="size",data=var,palette="Oranges_r",kind="boxen",hue="sex")
plt.show()
```
