```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var= pd.read_csv(r'C:\Users\User\juypter\files\tips.csv')
var
```

```python
sns.set(style="whitegrid")

sns.boxenplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxenplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxenplot(x="day",y="total_bill",data=var,hue="sex",color="r",order=["Fri","Sun","Thur","Sat"],linewidth=0.3,
              palette="terrain")
plt.show()
```

```python
sns.set(style="whitegrid")

sns.boxenplot(y="day",x="total_bill",data=var,hue="sex",color="r",order=["Fri","Sun","Thur","Sat"],linewidth=0.3,
              palette="plasma")
plt.show()
```
