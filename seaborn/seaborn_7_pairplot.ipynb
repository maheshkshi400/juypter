```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var =pd.read_csv(r'C:\Users\User\juypter\files\tips.csv')
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],x_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],y_vars=["total_bill","tip"])
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],y_vars=["total_bill","tip"],kind="reg")
plt.show()
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],markers=["o","*"],y_vars=["total_bill","tip"])
plt.show()
```
