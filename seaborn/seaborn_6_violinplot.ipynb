```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var= pd.read_csv(r'C:\Users\User\juypter\files\tips.csv')
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=0.1,palette='turbo_r')
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",hue="sex",split=True,data=var,linewidth=0.2,palette='turbo_r',
               order=['Dinner','Lunch'],saturation=1)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,palette='turbo_r',linewidth=0.2,scale="width")
plt.show()
```

```python
sns.violinplot(x="total_bill",y="day",data=var,hue="sex",split=True,palette='turbo_r',linewidth=0.2,scale="width",
               inner="quart")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",hue="sex",split=True,data=var,linewidth=0.2,palette='turbo_r',
               order=['Dinner','Lunch'],saturation=1,inner="quart") # (quartile,point,stick,box)
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",hue="sex",split=True,data=var,linewidth=0.2,palette='turbo_r',
               order=['Dinner','Lunch'],saturation=1,inner="box") # (quartile,point,stick,box)
plt.show()
```
