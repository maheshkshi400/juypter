```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var= pd.read_csv(r'C:\Users\User\juypter\files\tips.csv')
var
```

```python
fg = sns.FacetGrid(var,col="sex",hue="day")

fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col="day",hue="sex")

fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col="day",hue="sex")

fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg = sns.FacetGrid(var,col="sex",hue="day",palette="turbo")

fg.map(plt.scatter,"total_bill","tip",edgecolor="r").add_legend()
plt.show()
```
