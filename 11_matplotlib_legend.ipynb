```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - markerfirst,shadow,frameon,columnspacing,title,title_fontsize,borderpad,labelspacing


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red",label="label1")
plt.plot(v**3,v,color="green",label="label2")
plt.plot(v,v**4,color="orange",label="label3")
plt.plot(v,v**5,color="magenta",label="label4")

plt.xlim(0,4)
plt.ylim(0,4)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()


plt.legend(loc=(1,1),bbox_to_anchor=(0.5,0.5,0.5,0.5),fontsize=14,facecolor="yellow",edgecolor="green",
           framealpha=True,markerfirst=False,shadow=True,frameon=True,
           ncol=2,columnspacing=1,title="My legend",title_fontsize=14,borderpad=1,labelspacing=1)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - markerfirst,shadow,frameon,columnspacing,title,title_fontsize,borderpad,labelspacing


v= np.array([1,2,3,4])
a, = plt.plot(v**2,v,color="red")
b, = plt.plot(v**3,v,color="green")
c, =plt.plot(v,v**4,color="orange")
d, =plt.plot(v,v**5,color="magenta")

plt.xlim(0,4)
plt.ylim(0,4)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()


legend1 = plt.legend((a,b),["label1","label2"])
plt.gca().add_artist(legend1)
plt.legend((c,d),["label3","label4"])


plt.legend(fontsize=14,facecolor="yellow",edgecolor="green",
           framealpha=True,markerfirst=False,shadow=True,frameon=True,
           ncol=2,columnspacing=1,title="My legend",title_fontsize=14,borderpad=1,labelspacing=1)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - multiple legend


v= np.array([1,2,3,4])
a,=plt.plot(v**2,v,color="red")
b,=plt.plot(v**3,v,color="green")
c,=plt.plot(v,v**4,color="orange")
d,=plt.plot(v,v**5,color="magenta")


legend1 =plt.legend((a,b),["label1","label2"],loc="lower right",fontsize=14,facecolor="yellow",edgecolor="green",
           framealpha=True,markerfirst=False,shadow=True,frameon=True,
           ncol=2,columnspacing=1,title="My legend 1",title_fontsize=14,borderpad=1,labelspacing=1)
plt.gca().add_artist(legend1)
plt.legend((c,d),["label3","label4"],fontsize=14,facecolor="yellow",edgecolor="green",
           framealpha=True,markerfirst=False,shadow=True,frameon=True,
           ncol=2,columnspacing=1,title="My legend 2",title_fontsize=14,borderpad=1,labelspacing=1)


plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()




plt.tight_layout()
plt.show()
```
