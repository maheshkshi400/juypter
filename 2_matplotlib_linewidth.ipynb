```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(0,10,0.5)
plt.plot(a,"go:",a**2,"r^--",a**3,"c+-.")
plt.plot([10,20,30,40],[1,2,3,4],'kd-')

plt.xlabel("x label")
plt.ylabel("y lable")
plt.title("first plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.arange(0,10,0.5)

# ue alpha for transperancy of color

plt.plot([10,20,30,40],[1,2,3,4],alpha=0.7)

plt.xlabel("x label")
plt.ylabel("y lable")
plt.title("first plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# zorder for showing the graph front or back 


plt.plot([5,15,45,76,89],alpha=0.7,linewidth=15,zorder=2)
plt.plot([10,20,30,40,50,60,70],alpha=0.7,linewidth=15,zorder=1)

plt.xlabel("x label")
plt.ylabel("y lable")
plt.title("first plot")
plt.show()
```
