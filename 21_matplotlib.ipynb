```python
import matplotlib.pyplot as plt
import numpy as np

w=0.3

x=["MECH","PRODUCTION","EEE","AERONAUTICS","CIVIL","BIOTECH"]

boys =[120,320,200,240,120,320]
girls= [320,160,220,170,220,150]

bar1 = np.arange(len(x))
bar2=[i+w for i in bar1]

plt.bar(bar1,boys,w,label="boys")
plt.bar(bar2,girls,w,label="girls")

plt.xlabel("courses")
plt.ylabel("stdents")
plt.title("Students Vs Courses")
plt.xticks(bar1+w/2,x)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# types of plot-horizontal bar graph

x=["Russia","china","argentina","chillie","pakistan","Bharat"]


gold=[10,40,70,90,120,40]
silver=[108,200,205,107,302,204]
bronze=[110,107,103,127,105,310]


bar1 = np.arange(len(x))
bar2=[i+w for i in bar1]
bar3=[i+w for i in bar2]


plt.bar(bar1,gold,0.3,label="gold",color="red",edgecolor="black")
plt.bar(bar2,silver,0.3,label="silver",color="yellow",edgecolor="black")
plt.bar(bar3,bronze,0.3,label="bronze",color="green",edgecolor="black")


plt.xlabel("countries")
plt.ylabel("medals")
plt.title("countries vs medals")
plt.xticks(bar1+w,x)
plt.legend()

plt.show()
```
