```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'Sam Walton'
    return cell

def convert_price_cell(cell):
    if cell=="n.a.":
        return 50
    return cell
    
df=pd.read_excel(r'C:\Users\User\juypter\files\stock_data12.xlsx',converters= {
        'people': convert_people_cell,
        'price': convert_price_cell
    })


df
```

```python
df.to_excel(r'C:\Users\User\juypter\files\new.xlsx', sheet_name="stocks", index=False, startrow=2, startcol=1)
```
