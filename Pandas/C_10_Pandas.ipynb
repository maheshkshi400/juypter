```python
# feature scaling technique
# two types of feature scaling techniques
# 1. standardization
# 2. normalization
# standardization
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\User\juypter\files\loan.csv')
data.head(3)
```

```python
data.isnull().sum()
```

```python
data["CoapplicantIncome"].fillna(data["CoapplicantIncome"].median(), inplace=True)
```

```python
import matplotlib.pyplot as plt
sns.distplot(data["ApplicantIncome"])
plt.show()
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(data[["ApplicantIncome"]])
```

```python
data["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(data[["ApplicantIncome"]]),columns=["x"])
```

```python
data.head(3)
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 7))

plt.subplot(1,2,1)
plt.title("BEFORE")
sns.distplot(data["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("AFTER")
sns.distplot(data["ApplicantIncome_ss"])

plt.show()
```
