```python
import pandas as pd
var = pd.DataFrame({"days":[1,2,3,4,5,6],
                    "std_name":["suraj","Akshay","shaurya","suraj","Akshay","shaurya"],
                    "eng":[10,17,18,15,14,11],
                    "Maths":[9,12,15,17,19,20]})
var
```

```python
df= var.pivot(index="days",columns="std_name")
df
```

```python
df= var.pivot(index="days",columns="std_name",values="eng")
df
```

```python
import pandas as pd
var = pd.DataFrame({"days":[1,1,1,1,2,2],
                    "std_name":["suraj","Akshay","shaurya","suraj","Akshay","shaurya"],
                    "eng":[10,17,18,15,14,11],
                    "Maths":[9,12,15,17,19,20]})
var
```

```python
var.pivot_table(index="std_name",columns="days",aggfunc="mean")
```

```python
var.pivot_table(index="std_name",columns="days",aggfunc="sum",margins=True)
```
