```python

# Creating databases- Dictionary


import pandas as pd
import numpy as np
weather_data = {
        'Day':['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
        'temprature':[32,33,40,41,15,20], 
        'windspeed':[2,4,6,3,7,5],
        'events':['rain','snow','sunny','sunny','snow','rain']
             
}
df1 = pd.DataFrame(weather_data)
df1
```

```python
# Creating dataframes- tuples


import pandas as pd
import numpy as np
weather_data = [
        ('1/1/2017',32,2,'rain'),
        ('1/2/2017',33,4,'snow'),
        ('1/3/2017',40,6,'sunny'),
        ('1/4/2017',41,3,'sunny'),
        ('1/5/2017',15,7,'snow'),
        ('1/6/2017',20,5,'rain')
        ]       
df1 = pd.DataFrame(weather_data,columns=["Day","temprature","windspeed","event"])
df1
```
