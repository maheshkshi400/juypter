```python
import matplotlib.pyplot as plt
import numpy as np

t= [0,30,45,60,90]

x = [i*(np.pi/180) for i in t]

plt.plot(x,np.sin(x),marker="+",label="sin")

plt.plot(x,np.cos(x),marker="^",label="cos")


plt.title("sin and cos graph")
plt.xlabel("angle")
plt.ylabel("sin and cos values")




plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t= np.linspace(-np.pi,np.pi)

x = [i*(np.pi/180) for i in t]

plt.plot(t,np.sin(t),marker="+",label="sin")
plt.plot(t,np.cos(t),marker="^",label="cos")

plt.title("sin and cos graph")
plt.xlabel("angle")
plt.ylabel("sin and cos values")
plt.legend()

plt.show()
```
