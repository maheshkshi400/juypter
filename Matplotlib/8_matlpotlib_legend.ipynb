```python
import matplotlib.pyplot as plt
import numpy as np

# ylim function


plt.plot([5,15,45,76,89],alpha=0.7,linewidth=15,zorder=2)
plt.plot([10,20,30,40,50,60,70],alpha=0.7,linewidth=15,zorder=1)

plt.ylim(bottom=0,top=100)

# change font style,font size,fontname
plt.xlabel("x label",{"size":15,"color":"green"},labelpad = 10,alpha=0.7,backgroundcolor="red",style="italic")
plt.ylabel("y lable",{"size":15,"color":"red","backgroundcolor":"blue"},labelpad =10,alpha=0.7,family="cursive")
plt.title("first plot",{"size":15,"color":"green"},loc="center",backgroundcolor="pink",pad=15)
plt.legend()

plt.minorticks_on()
plt.grid(which="minor",color="green",linewidth=0.5,axis="both")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - Methid 1 #

v= np.array([1,2,3,4])
plt.plot([5,15,45,76,89],alpha=0.7,linewidth=15,zorder=2)
plt.plot([10,20,30,40,50,60,70],alpha=0.7,linewidth=15,zorder=1)
plt.plot(v**2,v,color="red",label="square")
plt.plot(v**3,v,color="green",label="cubes")


# change font style,font size,fontname
plt.xlabel("xlabel",{"size":15,"color":"green"},labelpad = 10,alpha=0.7,backgroundcolor="red",style="italic")
plt.ylabel("ylable",{"size":15,"color":"red","backgroundcolor":"blue"},labelpad =10,alpha=0.7,family="cursive")
plt.title("first plot",{"size":15,"color":"green"},loc="center",backgroundcolor="pink",pad=15)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - Method 2

v= np.array([1,2,3,4])

plt.plot([5,15,45,76,89],alpha=0.7,linewidth=3,zorder=2)
plt.plot([10,20,30,40,50,60,70],alpha=0.7,linewidth=3,zorder=1)
plt.plot(v**2,v,color="red")
plt.plot(v**3,v,color="green")


# change font style,font size,fontname
plt.xlabel("xlabel",{"size":15,"color":"green","backgroundcolor":"blue"})
plt.ylabel("ylable",{"size":15,"color":"red","backgroundcolor":"blue"})
plt.title("first plot",{"size":15,"color":"green"},loc="center",backgroundcolor="pink",pad=15)
plt.legend(["squares","cubes","area","circles"])

plt.show()
```
