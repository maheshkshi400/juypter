```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - fontsize


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red",label="label1")
plt.plot(v**3,v,color="green",label="label2")
plt.plot(v,v**4,color="orange",label="label3")
plt.plot(v,v**5,color="magenta",label="label4")

plt.xlim(0,4)
plt.ylim(0,4)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()

plt.legend(loc=(1,1),bbox_to_anchor=(0.5,0.5,0.5,0.5))
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - fontsize,facecolor,edgecolor,framealpha


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red",label="label1")
plt.plot(v**3,v,color="green",label="label2")
plt.plot(v,v**4,color="orange",label="label3")
plt.plot(v,v**5,color="magenta",label="label4")

plt.xlim(0,4)
plt.ylim(0,4)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()

plt.legend(loc=(1,1),bbox_to_anchor=(0.5,0.5,0.5,0.5))
plt.legend(fontsize=14,facecolor="yellow",edgecolor="green",framealpha=1)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - markerfirst,shadow


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red",label="label1")
plt.plot(v**3,v,color="green",label="label2")
plt.plot(v,v**4,color="orange",label="label3")
plt.plot(v,v**5,color="magenta",label="label4")

plt.xlim(0,4)
plt.ylim(0,4)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()

plt.legend(loc=(1,1),bbox_to_anchor=(0.5,0.5,0.5,0.5))
plt.legend(fontsize=14,facecolor="yellow",edgecolor="green",framealpha=True,shadow=True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - markerfirst,shadow,frameon,columnspacing


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red",label="label1")
plt.plot(v**3,v,color="green",label="label2")
plt.plot(v,v**4,color="orange",label="label3")
plt.plot(v,v**5,color="magenta",label="label4")

plt.xlim(0,4)
plt.ylim(0,4)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()

plt.legend(loc=(1,1),bbox_to_anchor=(0.5,0.5,0.5,0.5))
plt.legend(fontsize=14,facecolor="yellow",edgecolor="green",framealpha=True,markerfirst=False,shadow=True,frameon=True,
           ncol=2,columnspacing=1)

plt.tight_layout()
plt.show()
```
