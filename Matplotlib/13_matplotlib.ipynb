```python
import matplotlib.pyplot as plt
import numpy as np

# function - params


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red")
plt.plot(v**3,v,color="green")



plt.xlim(0,10)
plt.ylim(0,10)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()
plt.minorticks_on()
plt.tick_params(width=5,length=10,color="blue",pad=10,direction="inout",which="both",
                right=True,top=True,left=False,bottom=False,labelsize=12,labelcolor="red",
                labeltop=True,labelright=True,labelrotation=30,grid_color="green",grid_alpha=1,
                grid_linewidth=0.5,grid_linestyle="--")

plt.show()
```
