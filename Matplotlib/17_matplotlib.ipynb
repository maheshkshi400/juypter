```python
import matplotlib.pyplot as plt
import numpy as np

# text function,annotation

plt.plot([10,20,30,40],marker="o")
plt.text(1.5,25,"point",bbox={"fc":"green","ec":"blue","boxstyle":"roundtooth"})
plt.text(0.5,15,"newpoint",bbox={"fc":"pink","ec":"red","boxstyle":"circle"})
# "arrowstyle":"<->"}

# boxstyle":"circle","darrow","larrow","rarrow","round","smooth","roun4","square","roundtooth","sawtooth"

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# text function,annotation

plt.plot([10,20,30,40],marker="o")
plt.annotate("point",(0,10),(0.5,25),size=10,color="green",arrowprops={"width":10},
             bbox={"fc":"cyan","ec":"blue","boxstyle":"roundtooth","pad":0.7})

# "arrowstyle":"<->"}

plt.show()
```
