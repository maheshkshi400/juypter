```python
import sympy as sp
```

```python
x = sp.Symbol('x')
f = input("Enter the (in the terms of x) : ")

F_indefinite = sp.integrate(f, x)
print("Indefinite Integral of  f(x) dx : ", F_indefinite)

a = float(input("Enter the lower limit of the integral : "))
b = float(input("Enter the upper limit of the integral : "))
F_definite = sp.integrate(f, (x, a, b))
print("Definite Integral of  f(x) dx : ", F_definite)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(10000)
y = np.random.randn(10000)

plt.hexbin(x, y, gridsize=30, cmap='Purples')
plt.colorbar(label='Density')

plt.title('Hexagonal Binning')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
