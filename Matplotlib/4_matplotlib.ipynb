```python
import matplotlib.pyplot as plt
import numpy as np

# zorder for showing the graph front or back 


plt.plot([5,15,45,76,89],alpha=0.7,linewidth=15,zorder=2)
plt.plot([10,20,30,40,50,60,70],alpha=0.7,linewidth=15,zorder=1)

# change color of lable,size and labelpad distance away from x axis size extra small
plt.xlabel("x label",{"size":15,"color":"green"},labelpad = 10,alpha=0.7,backgroundcolor="red")
plt.ylabel("y lable",{"size":15,"color":"red","backgroundcolor":"blue"},labelpad =10,alpha=0.7)
plt.title("first plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# zorder for showing the graph front or back 


plt.plot([5,15,45,76,89],alpha=0.7,linewidth=15,zorder=2)
plt.plot([10,20,30,40,50,60,70],alpha=0.7,linewidth=15,zorder=1)

# change font style,font size,fontname
plt.xlabel("x label",{"size":15,"color":"green"},labelpad = 10,alpha=0.7,backgroundcolor="red",style="italic")
plt.ylabel("y lable",{"size":15,"color":"red","backgroundcolor":"blue"},labelpad =10,alpha=0.7,family="cursive")
plt.title("first plot",style="oblique")
plt.show()
```
