```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - ncol

v= np.array([1,2,3,4])

plt.plot([5,15,45,76,89],alpha=0.7,linewidth=3,zorder=2)
plt.plot([10,20,30,40,50,60,70],alpha=0.7,linewidth=3,zorder=1)
plt.plot(v**2,v,color="red")
plt.plot(v**3,v,color="green")


# change font style,font size,fontname
plt.xlabel("xlabel",{"size":15,"color":"green","backgroundcolor":"blue"})
plt.ylabel("ylable",{"size":15,"color":"red","backgroundcolor":"blue"})
plt.title("first plot",{"size":15,"color":"green"},loc="center",backgroundcolor="pink",pad=15)
plt.legend(["squares","cubes","area","circles"],ncol=2)
plt.grid()
plt.legend(loc=(1.5,0.5))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - bbox

v= np.array([1,2,3,4])

plt.plot([5,15,45,76,89],alpha=0.7,linewidth=3,zorder=2)
plt.plot([10,20,30,40,50,60,70],alpha=0.7,linewidth=3,zorder=1)
plt.plot(v**2,v,color="red")
plt.plot(v**3,v,color="green")


# change font style,font size,fontname
plt.xlabel("xlabel",{"size":15,"color":"green","backgroundcolor":"blue"})
plt.ylabel("ylable",{"size":15,"color":"red","backgroundcolor":"blue"})
plt.title("first plot",{"size":15,"color":"green"},loc="center",backgroundcolor="pink",pad=15)
plt.legend(["squares","cubes","area","circles"],ncol=2)
plt.grid()
plt.grid()
plt.legend(loc= "upper right",bbox_to_anchor=(0.5,0.5,0.5,0.5))


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - bbox


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red",label="label1")
plt.plot(v**3,v,color="green",label="label2")
plt.plot(v,v**4,color="orange",label="label3")
plt.plot(v,v**5,color="magenta",label="label4")

plt.xlim(0,4)
plt.ylim(0,4)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()

plt.legend(loc= "center",bbox_to_anchor=(0.5,0.5,0.5,0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# legend function - bbox,loc,tight_layout


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red",label="label1")
plt.plot(v**3,v,color="green",label="label2")
plt.plot(v,v**4,color="orange",label="label3")
plt.plot(v,v**5,color="magenta",label="label4")

plt.xlim(0,4)
plt.ylim(0,4)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()

plt.legend(loc=(1,1),bbox_to_anchor=(0.5,0.5,0.5,0.5))
plt.tight_layout()
plt.show()
```
