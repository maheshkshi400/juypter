```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000,3)

plt.hist(x,30,histtype="barstacked")

plt.xlabel("age")
plt.ylabel("cont of the people")
plt.title("population count")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)
y=np.sin(x)
z=np.cos(x)

plt.hist(x,30,histtype="barstacked")

plt.hist(x,bins="auto",ec="k")
plt.hist(y,bins="auto",ec="k")
plt.hist(z,bins="auto",ec="k")

plt.xlabel("age")
plt.ylabel("cont of the people")
plt.title("population count")

plt.show()
```
