```python
import pandas as pd


data1 = {
    'id':[1,2,3,4,5],
    'Name':["mahesh","shruthi",'pooja','kamini','shivaji'],
    'gender':["male","female","female","female","male"],
    'age':[20,30,25,34,31],  # Adjusted to have 10 elements
    'salary':[5000,6000,7000,8000,6000],
    'dept':["IT","HR","FINANCE","PAYROLL","IT"],
    'deptno':[50,20,40,30,50]}

data2 = {
    'id':[6,7,8,9,10],
    'Name':['shruthi','suresh','shivani','prachi','kirthi'],
    'gender':["female","male","female","female","female"],
    'age':[31,20,30,25,34],  # Adjusted to have 10 elements
    'salary':[6000,8000,6000,10000,11000],
    'dept':["IT","IT","HR","FINANCE","PAYROLL"],
    'deptno':[50,50,20,40,30]}

df1 = pd.DataFrame(data1)
df2= pd.DataFrame(data2)

df3 = df1.join(df2, lsuffix='_df1', rsuffix='_df2')
df3

```

```python
d1 = pd.DataFrame({"A": [1, 2, 3, 4, 5], "B": [11, 21, 31, 41, 51]},index=["a","b","c","d","e"])
d2 = pd.DataFrame({"C": [10, 20, 30, 40, 50], "D": [71, 72, 73, 74, 75]},index=["a","b","c","d","e"])

d3=d1.join(d2)
d3
```

```python
d1 = pd.DataFrame({"A": [1, 2, 3, 4, 5], "B": [11, 21, 31, 41, 51]},index=["a","b","c","d","e"])
d2 = pd.DataFrame({"C": [10, 20, 30], "D": [71, 72, 73]},index=["a","b","c"])

d3=d1.join(d2,how='inner')
d3
```

```python
d1 = pd.DataFrame({"A": [1, 2, 3, 4, 5], "B": [11, 21, 31, 41, 51]},index=["a","b","c","d","e"])
d2 = pd.DataFrame({"C": [10, 20, 30], "D": [71, 72, 73]},index=["a","b","c"])

d3=d1.join(d2,how="outer",lsuffix="_12",rsuffix="_123")
d3
```
