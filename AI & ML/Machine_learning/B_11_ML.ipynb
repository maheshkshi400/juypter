```python
# feature scaling technique
# two types of feature scaling techniques
# 2. normalization

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\User\juypter\files\loan.csv')
data.head(3)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.distplot(data["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(data[["CoapplicantIncome"]])
```

```python
data["CoapplicantIncome_min"] = ms.transform(data[["CoapplicantIncome"]])
```

```python
data.head(3)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.title("BEFORE")
sns.distplot(data["CoapplicantIncome"])

plt.subplot(1, 2, 2)
plt.title("AFtER")
sns.distplot(data["CoapplicantIncome_min"])

plt.show()
```
