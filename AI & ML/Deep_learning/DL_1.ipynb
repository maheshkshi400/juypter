```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv(r"C:\Users\User\juypter\files\placement1.csv")
data.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x = "cgpa",y ="resume_score",data=data,hue="placed")
plt.show()
```

```python
x = data.iloc[:,:-1]
y = data["placed"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import Perceptron

pr = Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_test,y_test)*100,pr.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(3,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
plt.show()
```
