```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv(r"C:\Users\User\juypter\files\churn.csv")
data.head(3)
```

```python
data2 = data.drop(columns=['Surname','Geography','Gender','RowNumber','CustomerId'],axis=1)
```

```python
data2.shape
```

```python
input_data = data2.iloc[:,:-1]
output_data = data2.iloc[:,-1]
```

```python
ss = StandardScaler()
input_data =pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(input_data,output_data,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6,input_dim=8,activation="relu"))
ann.add(Dense(4,activation="relu"))
ann.add(Dense(2,activation="relu"))
ann.add(Dense(1,activation="sigmoid"))
```

```python
ann.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=100,epochs=50,validation_data=(x_test,y_test))
```

```python
prd_data = ann.predict(x_test)
```
