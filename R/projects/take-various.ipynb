```r
# Instalar e carregar a biblioteca necessária
install.packages("ggplot2")
library(ggplot2)

# Criar os dados
dados <- data.frame(
  Country = c("Sweden", "New Zealand", "Poland", "Mexico", "Norway", 
              "Australia", "Iceland", "Canada", "Czechia", "Denmark", 
              "France", "Germany", "Portugal", "South Korea", "Spain", 
              "Costa Rica", "Finland"),
  Year = c(2017, 2018, 2019, 2020, 2020, 2021, 2021, 2022, 2022, 2022, 
           2022, 2022, 2022, 2022, 2022, 2023, 2023),
  Value1 = c(64328856, 42239723, 664557, 45213995, 12268863, 
             2973753, 52077324, 2389166, 85877396, 905157, 
             809507, 45943825, 0, 0, 15282989, 0, 0),
  Value2 = c(4892771, 13075949, 4876978, 31673275, 24723616, 
             945497, 0, 8068862, 19842883, 54190987, 
             29878168, 21253494, 0, 0, 38993984, 0, 0)
)

# Criar o gráfico de linhas
ggplot(dados, aes(x = Country, y = Value1, group = 1)) +
  geom_line(color = "purple", size = 1) +
  geom_point(color = "purple", size = 3) +
  theme_minimal() +
  labs(title = "Gráfico de Linhas Roxo",
       x = "País",
       y = "Valor 1") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
