```r
library(ggplot2)
library(maps)

# Load world map data
world_map <- map_data("world")

# Draw the world map
ggplot(world_map, aes(x = long, y = lat, group = group)) +
  geom_polygon(fill = "lightblue", color = "black") +
  theme_minimal() +
  labs(title = "World Map",
       x = "Longitude",
       y = "Latitude")
```

```r
# Install and load necessary packages
install.packages("ggplot2")
install.packages("maps")
install.packages("dplyr")

library(ggplot2)
library(maps)
library(dplyr)

# Load world map data
world_map <- map_data("world")

# Highlight India by assigning it a unique fill color
ggplot() +
  geom_polygon(data = world_map, aes(x = long, y = lat, group = group, fill = region == "India"), 
               color = "black") +
  scale_fill_manual(values = c("lightblue", "orange"), guide = FALSE) +
  theme_minimal() +
  labs(title = "World Map Highlighting India",
       x = "Longitude",
       y = "Latitude")
```

```r
library(ggplot2)
library(rnaturalearth)
library(sf)
library(dplyr)
```

```r
# Load India map data
india <- ne_states(country = "India", returnclass = "sf")

# Example biodiversity data (e.g., species richness by state)
biodiversity_data <- data.frame(
  state = india$name, # State names from the map data
  richness = runif(n = nrow(india), min = 50, max = 200) # Random species richness values
)

# Merge map and biodiversity data
india_biodiversity <- india %>%
  left_join(biodiversity_data, by = c("name" = "state"))
```

```r
# Plotting the India map with biodiversity data
ggplot(data = india_biodiversity) +
  geom_sf(aes(fill = richness), color = "black") +
  scale_fill_viridis_c(option = "plasma", name = "Species Richness") +
  theme_minimal() +
  labs(title = "Biodiversity in India: Species Richness by State",
       subtitle = "Sample data for demonstration",
       x = "Longitude",
       y = "Latitude") +
  theme(legend.position = "right")
```

```r
# Load China map data
china <- ne_states(country = "China", returnclass = "sf")

# Example biodiversity data (e.g., species richness by province)
biodiversity_data <- data.frame(
  province = china$name, # Province names from the map data
  richness = runif(n = nrow(china), min = 50, max = 200) # Random species richness values
)

# Merge map and biodiversity data
china_biodiversity <- china %>%
  left_join(biodiversity_data, by = c("name" = "province"))

# Plotting the China map with biodiversity data
ggplot(data = china_biodiversity) +
  geom_sf(aes(fill = richness), color = "black") +
  scale_fill_viridis(option = "plasma", name = "Species Richness") +
  theme_minimal() +
  labs(title = "Biodiversity in China: Species Richness by Province",
       subtitle = "Sample data for demonstration",
       x = "Longitude",
       y = "Latitude") +
  theme(legend.position = "right")
```
