```r
# Assuming your data is stored in the 'iris' dataset
data(iris)
```

```r
# Calculate the count of each species
species_count <- table(iris$Species)

# Calculate relative abundance
relative_abundance <- prop.table(species_count)

# Print the relative abundance
relative_abundance
```

```r
# Convert relative abundance to a data frame
abundance_df <- as.data.frame(relative_abundance)

# Plot the relative abundance
library(ggplot2)
ggplot(abundance_df, aes(x = Var1, y = Freq, fill = Var1)) +
  geom_bar(stat = "identity") +
  labs(title = "Relative Abundance of Species",
       x = "Species",
       y = "Relative Abundance") +
  theme_minimal()
```

```r
# Assuming 'iris' dataset

# Create a stacked bar plot
ggplot(iris, aes(x = Species, fill = Species)) +
  geom_bar(position = "fill") +
  labs(title = "Stacked Bar Plot of Relative Abundance",
       x = "Species",
       y = "Relative Abundance") +
  scale_y_continuous(labels = scales::percent) +
  theme_minimal()
```

```r
# Calculate relative abundance
species_count <- table(iris$Species)
relative_abundance <- prop.table(species_count)

# Convert to a data frame
abundance_df <- as.data.frame(relative_abundance)

# Plot pie chart
ggplot(abundance_df, aes(x = "", y = Freq, fill = Var1)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar("y") +
  labs(title = "Pie Chart of Relative Abundance") +
  theme_void() +
  theme(legend.title = element_blank())
```

```r
# Doughnut chart
ggplot(abundance_df, aes(x = 2, y = Freq, fill = Var1)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar(theta = "y") +
  xlim(0.5, 2.5) +
  labs(title = "Doughnut Chart of Relative Abundance") +
  theme_void() +
  theme(legend.title = element_blank())
```

```r
# Bubble chart
ggplot(abundance_df, aes(x = Var1, y = Freq, size = Freq, fill = Var1)) +
  geom_point(shape = 21, color = "black") +
  labs(title = "Bubble Chart of Relative Abundance",
       x = "Species",
       y = "Relative Abundance") +
  theme_minimal()
```

```r
# Example of a heatmap (for multiple samples)

# Create an example abundance matrix
species_matrix <- table(iris$Species, iris$Sepal.Width)
heatmap(as.matrix(species_matrix),
        col = colorRampPalette(c("white", "blue"))(50),
        scale = "row",
        main = "Heatmap of Relative Abundance",
        xlab = "Sepal Width",
        ylab = "Species")
```
