 # t-Test and ANOVA 


```r
# Filter the dataset to include only "setosa" and "versicolor"
iris_filtered <- subset(iris, Species %in% c("setosa", "versicolor"))
```

```r
# Perform a t-test
t_test_result <- t.test(Sepal.Length ~ Species, data = iris_filtered)

# Print the result
print(t_test_result)
```

```r
# ANOVA (Analysis of Variance)

# Perform ANOVA
anova_result <- aov(Sepal.Length ~ Species, data = iris)

# Print the summary of the ANOVA
summary(anova_result)
```

```r
# Perform Tukey's HSD test
tukey_result <- TukeyHSD(anova_result)

# Print the result
print(tukey_result)

# Plot the Tukey's HSD result
plot(tukey_result)
```

```r
# Boxplot for Sepal.Length across species
boxplot(Sepal.Length ~ Species, data = iris,
        main = "Boxplot of Sepal Length by Species",
        xlab = "Species",
        ylab = "Sepal Length",
        col = "lightblue")
```
