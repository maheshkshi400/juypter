```r
library(ggplot2)

# Multiple linear regression
model <- lm(cbind(Sepal.Length, Sepal.Width, Petal.Length, Petal.Width) ~ Species, data = iris)

# Extract coefficients
coefs <- summary(model)$coefficients

# Create a scatter plot with regression lines for each response variable 
ggplot(iris, aes(x = Species)) +
  geom_point(aes(y = Sepal.Length), color = "blue") +
  geom_hline(yintercept = coefs[1, 1], color = "blue", linetype = "dashed") +
  geom_point(aes(y = Sepal.Width), color = "green") +
  geom_hline(yintercept = coefs[2, 1], color = "green", linetype = "dashed") +
  geom_point(aes(y = Petal.Length), color = "red") +
  geom_hline(yintercept = coefs[3, 1], color = "red", linetype = "dashed") +
  geom_point(aes(y = Petal.Width), color = "purple") +
  geom_hline(yintercept = coefs[4, 1], color = "purple", linetype = "dashed") +
  labs(title = "Regression Lines for Species vs. Variables")
```

```r
# Load necessary libraries
library(ggplot2)

# Load the iris dataset
data(iris)

# Perform linear regression
model <- lm(Sepal.Length ~ Petal.Width, data = iris)

# Print summary of the regression model
summary(model)

# Create scatter plot with regression line
ggplot(iris, aes(x = Petal.Width, y = Sepal.Length)) +
  geom_point(color = "blue") +
  geom_smooth(method = "lm", color = "red") +
  labs(title = "Linear Regression: Sepal Length vs Petal Width",
       x = "Petal Width",
       y = "Sepal Length") +
  theme_minimal()
```

```r
# Polynomial regression (degree 2)
poly_model <- lm(Sepal.Length ~ poly(Petal.Width, 2), data = iris)
summary(poly_model)

# Plot
ggplot(iris, aes(x = Petal.Width, y = Sepal.Length)) +
  geom_point(color = "blue") +
  geom_smooth(method = "lm", formula = y ~ poly(x, 2), color = "green") +
  labs(title = "Polynomial Regression (Degree 2): Sepal Length vs Petal Width",
       x = "Petal Width",
       y = "Sepal Length") +
  theme_minimal()
```

```r
# Polynomial regression (degree 3)
poly_model3 <- lm(Sepal.Length ~ poly(Petal.Width, 3), data = iris)
summary(poly_model3)

# Plot
ggplot(iris, aes(x = Petal.Width, y = Sepal.Length)) +
  geom_point(color = "blue") +
  geom_smooth(method = "lm", formula = y ~ poly(x, 3), color = "purple") +
  labs(title = "Polynomial Regression (Degree 3): Sepal Length vs Petal Width",
       x = "Petal Width",
       y = "Sepal Length") +
  theme_minimal()
```

```r
# Load necessary libraries
library(rpart)
library(rpart.plot)

# Decision tree regression
tree_model <- rpart(Sepal.Length ~ Petal.Width, data = iris)
summary(tree_model)

# Plot the decision tree
rpart.plot(tree_model)

# Plot predictions vs actual data
iris$predicted_tree <- predict(tree_model)
ggplot(iris, aes(x = Petal.Width, y = Sepal.Length)) +
  geom_point(color = "blue") +
  geom_line(aes(y = predicted_tree), color = "orange") +
  labs(title = "Decision Tree Regression: Sepal Length vs Petal Width",
       x = "Petal Width",
       y = "Sepal Length") +
  theme_minimal()
```

```r
# Load necessary libraries
library(e1071)

# Support Vector Regression (SVR)
svr_model <- svm(Sepal.Length ~ Petal.Width, data = iris)
summary(svr_model)

# Plot predictions vs actual data
iris$predicted_svr <- predict(svr_model)
ggplot(iris, aes(x = Petal.Width, y = Sepal.Length)) +
  geom_point(color = "blue") +
  geom_line(aes(y = predicted_svr), color = "darkred") +
  labs(title = "Support Vector Regression: Sepal Length vs Petal Width",
       x = "Petal Width",
       y = "Sepal Length") +
  theme_minimal()
```

```r
# Load necessary libraries
library(randomForest)

# Random Forest Regression
rf_model <- randomForest(Sepal.Length ~ Petal.Width, data = iris)
summary(rf_model)

# Plot predictions vs actual data
iris$predicted_rf <- predict(rf_model)
ggplot(iris, aes(x = Petal.Width, y = Sepal.Length)) +
  geom_point(color = "blue") +
  geom_line(aes(y = predicted_rf), color = "darkgreen") +
  labs(title = "Random Forest Regression: Sepal Length vs Petal Width",
       x = "Petal Width",
       y = "Sepal Length") +
  theme_minimal()
```
