```r
numerical_cols <- c("HH_Size")
find_outliers <- function(df, col) {
  Q1 <- quantile(df[[col]], 0.25, na.rm = TRUE)
  Q3 <- quantile(df[[col]], 0.75, na.rm = TRUE)
  IQR <- Q3 - Q1
  lower_bound <- Q1 - 1.5 * IQR
  upper_bound <- Q3 + 1.5 * IQR
  print(Q1)
  print(Q3)
  df %>% filter(df[[col]] < lower_bound | df[[col]] > upper_bound)
}

# Function to remove outliers but limit to 10% of data
remove_outliers_limited <- function(df, col, max_removal_fraction = 0.10) {
  outliers <- find_outliers(df, col)
  num_outliers <- nrow(outliers)
  total_rows <- nrow(df)
  
  if (num_outliers / total_rows > max_removal_fraction) {
    num_to_remove <- ceiling(max_removal_fraction * total_rows)
    outliers <- outliers %>% slice_head(n = num_to_remove)
  }
  
  df %>% anti_join(outliers, by = col)
}
 data <- remove_outliers_limited(data,"HH_Size")
data
```

```r
calculate_percentage_within_ranges <- function(actual, predicted, ranges = c(0.05, 0.1, 0.15, 0.2, 0.25)) {
  # Initialize a list to store percentages for each range
  percentages <- setNames(numeric(length(ranges)), paste0("±", ranges * 100, "%"))
  
  # Calculate the percentage for each range
  for (range in ranges) {
    lower_bound <- actual * (1 - range)
    upper_bound <- actual * (1 + range)
    
    within_range <- (predicted >= lower_bound) & (predicted <= upper_bound)
    count_within_range <- sum(within_range, na.rm = TRUE)
    
    percentages[paste0("±", range * 100, "%")] <- (count_within_range / length(actual)) * 100
  }
  
  return(percentages)
}

# Example usage with 'test_data'
percentages_within_ranges <- calculate_percentage_within_ranges(test_data$Yield, test_data$predicted)
print(percentages_within_ranges)
```
