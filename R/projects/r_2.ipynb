```r
library(ggplot2)

ggplot(data = iris, aes(x = Petal.Width, y = Sepal.Length)) + geom_point()
```

```r
# Load necessary libraries
library(ggplot2)
library(reshape2)

# Assuming your data is in a data frame called 'df'
df_melt <- melt(iris, id.vars = "Species")

# Check the reshaped data
head(df_melt)
```

```r
ggplot(df_melt, aes(x = Species, y = value, fill = variable)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(x = "Species", y = "Value", fill = "Variable") +
  theme_minimal()
```

```r
# Load necessary library
library(ggplot2)

# Create the bar plot
qplot(data = iris, x = Petal.Width, geom = "bar", fill = Species, col = Species) +
  labs(x = "Petal Width", y = "Count", fill = "Species", col = "Species") +
  theme_minimal()
```
