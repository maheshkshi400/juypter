# correlation plot

```r
# psych,corrplot

library(corrplot)

data <- iris
corr_matrix <- cor(data[-5])

corrplot(corr_matrix, method = "circle")
```

```r
# Assuming you've already installed and loaded the ggplot2 package
library(ggplot2)

# Create the bar plot
ggplot(data, aes(x = Species, y = Petal.Width)) +
  geom_bar(stat = "summary", fun = "mean") +
  labs(title = "Average Petal Width by Species",
       x = "Species",
       y = "Average Petal Width")
```

```r
# Load necessary libraries
library(ggplot2)

# Assuming your data frame is named 'df'
# Create a bar plot showing the mean Petal.Width for each species
ggplot(iris, aes(x = Species, y = Petal.Width)) +
  stat_summary(fun = mean, geom = "bar", fill = "skyblue", color = "black") +
  labs(title = "Mean Petal Width by Species", x = "Species", y = "Petal Width (Mean)") +
  theme_minimal()
```

```r
# Assuming you've already installed and loaded the ggplot2 package
library(ggplot2)

# Create the bar plot with different colors for each species
ggplot(data, aes(x = Species, y = Petal.Width, fill = Species)) +
  geom_bar(stat = "summary", fun = "mean") +
  labs(title = "Average Petal Width by Species",
       x = "Species",
       y = "Average Petal Width")
```
