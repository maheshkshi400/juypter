# Regression Analysis

```r
# Load the dataset
data(iris)
```

```r
# Simple linear regression
model <- lm(Sepal.Length ~ Petal.Width, data = iris)

# Print the summary of the model
summary(model)
```

```r
# Plot the data
plot(iris$Petal.Width, iris$Sepal.Length, main = "Linear Regression: Sepal Length vs Petal Width",
     xlab = "Petal Width", ylab = "Sepal Length", pch = 16, col = "blue")

# Add the regression line
abline(model, col = "red", lwd = 2)
```

```r
# Simple linear regression
model_simple <- lm(Sepal.Length ~ Petal.Width, data = iris)

# Plot the data with the regression line
plot(iris$Petal.Width, iris$Sepal.Length, main = "Simple Linear Regression: Sepal Length vs Petal Width",
     xlab = "Petal Width", ylab = "Sepal Length", pch = 16, col = "blue")
abline(model_simple, col = "red", lwd = 2)
```

```r
# Multiple linear regression
model_mult <- lm(Sepal.Length ~ Petal.Width + Petal.Length + Sepal.Width, data = iris)

# Print the summary of the model
summary(model_mult)
```

```r
# Multiple linear regression
model_mult <- lm(Sepal.Length ~ Petal.Width + Petal.Length + Sepal.Width, data = iris)

# 3D scatter plot for multiple regression
if (!requireNamespace("scatterplot3d", quietly = TRUE)) {
    install.packages("scatterplot3d")
}
library(scatterplot3d)

# Create the 3D plot
scatterplot3d(iris$Petal.Width, iris$Petal.Length, iris$Sepal.Length,
              main = "Multiple Linear Regression: Sepal Length vs Petal Width and Length",
              xlab = "Petal Width", ylab = "Petal Length", zlab = "Sepal Length", color = "blue",
              pch = 16)

# Add the regression plane
plane <- predict(model_mult, iris)
scatterplot3d(iris$Petal.Width, iris$Petal.Length, plane, type = "h", col = "red")
```

```r
# Polynomial regression (quadratic)
model_poly <- lm(Sepal.Length ~ poly(Petal.Width, 2), data = iris)

# Print the summary
summary(model_poly)

# Plot the data with the polynomial fit
plot(iris$Petal.Width, iris$Sepal.Length, main = "Polynomial Regression: Sepal Length vs Petal Width",
     xlab = "Petal Width", ylab = "Sepal Length", pch = 16, col = "blue")

# Add the fitted curve
lines(sort(iris$Petal.Width), fitted(model_poly)[order(iris$Petal.Width)], col = "red", lwd = 2)
```

```r
# Polynomial regression (quadratic)
model_poly <- lm(Sepal.Length ~ poly(Petal.Width, 2), data = iris)

# Plot the data
plot(iris$Petal.Width, iris$Sepal.Length, main = "Polynomial Regression: Sepal Length vs Petal Width",
     xlab = "Petal Width", ylab = "Sepal Length", pch = 16, col = "blue")

# Add the polynomial fit line
lines(sort(iris$Petal.Width), predict(model_poly)[order(iris$Petal.Width)], col = "red", lwd = 2)
```

```r
# Install necessary package
if (!requireNamespace("nnet", quietly = TRUE)) {
    install.packages("nnet")
}
library(nnet)

# Perform multinomial logistic regression
model_logistic <- multinom(Species ~ Petal.Width + Petal.Length, data = iris)

# Create a grid for prediction
grid <- expand.grid(
  Petal.Width = seq(min(iris$Petal.Width), max(iris$Petal.Width), length.out = 100),
  Petal.Length = seq(min(iris$Petal.Length), max(iris$Petal.Length), length.out = 100)
)

# Predict the species on the grid
grid$pred <- predict(model_logistic, newdata = grid)

# Plot the decision boundary
plot(iris$Petal.Width, iris$Petal.Length, col = as.numeric(iris$Species), pch = 16,
     xlab = "Petal Width", ylab = "Petal Length", main = "Logistic Regression Decision Boundary")
points(grid$Petal.Width, grid$Petal.Length, col = as.numeric(grid$pred), pch = 20, cex = 0.1)
legend("topright", legend = levels(iris$Species), col = 1:3, pch = 16)
```
