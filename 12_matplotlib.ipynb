```python
import matplotlib.pyplot as plt
import numpy as np

# function - xticks,yticks


v= np.array([1,2,3,4])
a,=plt.plot(v**2,v,color="red")
b,=plt.plot(v**3,v,color="green")
c,=plt.plot(v,v**4,color="orange")
d,=plt.plot(v,v**5,color="magenta")


plt.xlim(0,5)
plt.ylim(0,5)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()
print(plt.xticks([2,4],["A","B"]))
print(plt.yticks([1,5],["c","d"]))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# function - xticks,yticks


v= np.array([1,2,3,4])
plt.plot(v**2,v,color="red")
plt.plot(v**3,v,color="green")



plt.xlim(0,10)
plt.ylim(0,10)

plt.xlabel("x lable")
plt.ylabel("y lable")
plt.title("First plot")
plt.grid()

print(plt.xticks(np.arange(1,21,step=5),["A","B","C","D"],color="red",size="medium",rotation="vertical"))     # Method 1
print(plt.xticks([2,4],["A","B"],color="red",size="medium",rotation="vertical"))            # Method 2

print(plt.yticks(np.arange(1,21,step=5),["A","B","C","D"],color="red",size="medium",rotation="vertical"))     # Method 1
print(plt.yticks([2,4],["A","B"],color="red",size="medium",rotation="vertical"))  

plt.show()
```
