# SVM REGRESSION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
data = pd.read_csv(r'C:\Users\User\juypter\files\placement.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,:-1]
x.ndim
y=data["package"]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,3))
plt.style.use('ggplot')
sns.scatterplot(x="cgpa", y="package", data=data,hue="package")
plt.title("Scatterplot of CGPA vs Package")  
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
# 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'

sv = SVR(kernel='rbf')
sv.fit(x_train, y_train)
```

```python
sv.score(x_train, y_train)*100,sv.score(x_test, y_test)*100
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x="cgpa", y="package", data=data)
plt.plot(data["cgpa"],sv.predict(x),color="red",linewidth=2)
plt.title("Scatterplot of CGPA vs Package")  
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```

# using PLOY KERNEL

```python
from sklearn.svm import SVR

sv = SVR(kernel='poly',degree=7)
sv.fit(x_train, y_train)
```

```python
sv.score(x_train, y_train)*100,sv.score(x_test, y_test)*100
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x="cgpa", y="package", data=data)
plt.plot(data["cgpa"],sv.predict(x),color="red",linewidth=2)
plt.title("Scatterplot of CGPA vs Package")  
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.show()
```
