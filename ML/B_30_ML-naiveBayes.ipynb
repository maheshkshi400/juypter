# NAIVE BAYES ALGORITHM

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv(r'C:\Users\User\juypter\files\placement1.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
sns.kdeplot(data=data["cgpa"])
plt.title("KDE Plot of CGPA")
plt.show()
```

```python
import matplotlib.pyplot as plt
sns.kdeplot(data=data["resume_score"])
plt.title("KDE Plot of CGPA")
plt.show()
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 3))
sns.scatterplot(x="cgpa",y="resume_score",data=data,hue="placed")
plt.show()
```

```python
x = data.iloc[:,:-1]
x.ndim
y = data["placed"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

# Gaussian Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
gnb.score(x_test,y_test)*100
```

```python
gnb.score(x_train,y_train)*100
```

# Multinomial Naive Bayes

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
mnb.score(x_test, y_test)*100
```

```python
mnb.score(x_train, y_train)*100
```

```python
gnb.predict([[6.17,5.17]])
```

# Bernoulli Naive Bayes

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_test,y_test)*100
```

```python
bnb.score(x_train,y_train)*100
```
