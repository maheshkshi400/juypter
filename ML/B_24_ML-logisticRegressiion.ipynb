```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\User\juypter\files\Social_Network_Ads.csv')
data.head()
```

```python
data.drop(columns=["EstimatedSalary"],inplace=True)
```

```python
data
```

```python

plt.figure(figsize=(4,3))
sns.scatterplot(x="Age", y="Purchased", data=data)
plt.show()
```

```python
print(data.columns)
```

```python
x = data[["Age"]]
y = data["Purchased"]
```

```python
from sklearn.model_selection import train_test_split


x_test, x_train, y_test, y_train = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sympy.abc import y

lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[40]])
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age", y="Purchased", data=data)
sns.lineplot(x="Age", y=lr.predict(x), data=data, color="red")
plt.show()
```
