```python
import matplotlib.pyplot as plt
import numpy as np

# types of plot-line 

year=[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
unemployment_rate=[9.8,7,12,3.9,7,6.5,8,9.5,11.5,3.2]

plt.plot(year,unemployment_rate,marker="o")
plt.xlabel("year")
plt.ylabel("unemployment_rate")
plt.title("years vs unemployment_rate")

plt.show()

```
