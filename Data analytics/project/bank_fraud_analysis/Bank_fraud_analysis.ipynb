```python
data['Age']=data['Age'].astype(int)
```

```python
duplicated = data.duplicated().sum()
print("No. of Duplicated values:",duplicated)
```

```python
# Rmoving outliers  in "Annula income" column using IQR

Q1 = data["Annual_Income"].quantile(0.25)
Q3 = data["Annual_Income"].quantile(0.75)
IQR = Q3-Q1
df = data[(data['Annual_Income'] >= (Q1 - 1.5 * IQR)) & (data['Annual_Income'] <= (Q3 + 1.5 * IQR))]
```

```python
 # creating new column "income per account"
 
data["Income_per_account"]=data["Annual_Income"]/data["Num_Bank_Accounts"]
```

```python
data.head(5)
```

```python
# creating new column "Credit_score_level"

data['Credit_score_level'] = data['Credit_Score'].apply(
    lambda x: 'High' if isinstance(x, (int, float)) and x > 750 else 'Low'
)
```

KPI'S

```python
# Customer demographics:
# avg_age of customers
Avg_age = data["Age"].mean()
print("Average age of customers:",Avg_age)

#Distribution of Occupation
occupation_count = data['Occupation'].value_counts()
print("occupation count is:",occupation_count)
occupation_unique = data['Occupation'].nunique()
print("occupation count is:",occupation_unique )

# Number of customers  by income group

#Number of Customers by Income Group
#Defining income group (bins)
max_income = data['Annual_Income'].max()
bins = [0, 25000, 50000, 100000, 200000]
labels = ['Low (<25K)', 'Lower-Middle (25K - 50K)', 'Middle (50K-100K)', 'Upper-Middle (100K-200K)']

if max_income > 200000:
    print("Above Upper-Middle Group")

data['Income_Group'] = pd.cut(
    data['Annual_Income'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

#Number of customers by income group
income_group_counts = data['Income_Group'].value_counts().sort_index()
print(income_group_counts)



```

Finnancial Behaviour

```python
# Avg_anuual_Income
Avg_Annual_Income = data["Annual_Income"].mean()
print("Average annual income is:",Avg_Annual_Income)

# Avg_monthly_in_hand_salary
Avg_monthly_salary = data["Monthly_Inhand_Salary"].mean()
print("Avg monthly in hand salary is :",Avg_monthly_salary)

# Average Debt-to-income ratio
Debt_to_income_ratio = data["Outstanding_Debt"]/data["Annual_Income"]
Avg_Debt_to_income_ratio = Debt_to_income_ratio.mean()
print("Avg Debt to income ratio is: ",Avg_Debt_to_income_ratio)

# Average_credit_utilisation_ratio
credit_utilisation_ratio = data["Credit_Utilization_Ratio"]
Avg_credit_utilisation_ratio = credit_utilisation_ratio.mean()
print("Avg credit utilisation ratio is :",Avg_credit_utilisation_ratio)

# Avg_num_of_bank_Accounts
Avg_num_of_bank_Accounts = data["Num_Bank_Accounts"].mean()
print("Avg num of bank Accounts is :",Avg_num_of_bank_Accounts)

# Avg_no_of_credit_cards
Avg_no_of_credit_cards = data["Num_Credit_Card"].mean()
print("Avg no of credit cards is :",Avg_no_of_credit_cards)



```

Loans & Credit Performance

```python
# Avg_no_of_loans_per_customer
Avg_no_of_loans_per_customer =data["Num_of_Loan"].mean()
print("Avg no of loans per customer is:",Avg_no_of_loans_per_customer)

# Avg. interest rate
Avg_interest_rate = data["Interest_Rate"].mean()
print("Average interest rate is :",Avg_interest_rate)

# Total Emi as % ofincome(EMI Burden Ratio)
total_EMI = data["Total_EMI_per_month"].sum()
total_income= data["Annual_Income"].sum()
EMI_Burden_Ratio =(total_EMI/total_income)*100
print("Total Emi as % ofincome(EMI Burden Ratio):",EMI_Burden_Ratio)

# Average Credit History Age
Average_Credit_History_Age =data["Credit_History_Age"].mean()
print("Average Credit History Age is: ",Average_Credit_History_Age)
```

Risk & Fraud Indicators

```python
# Avg_no_of_delayed_payments
Avg_no_of_delayed_payments = data["Num_of_Delayed_Payment"].mean()
print("Avg_no_of_delayed_payments is:",Avg_no_of_delayed_payments)

# Delay_from_Due_date (Avg,min,max)
Delay_from_Due_date_min = data["Delay_from_due_date"].min()
print("Delay_from_Due_date(min):",Delay_from_Due_date_min)

Delay_from_Due_date_max = data["Delay_from_due_date"].max()
print("Delay_from_Due_date(max):",Delay_from_Due_date_max)

Delay_from_Due_date_Avg = data["Delay_from_due_date"].mean()
print("Delay_from_Due_date(Avg):",Delay_from_Due_date_Avg)

# percentage of customers with frequent Credit inquries
percent_of_customers_with_frequent_Credit_inquries = (data["Num_Credit_Inquiries"]>0).mean()*100
print("percent of customers with frequent Credit_ inquries:",percent_of_customers_with_frequent_Credit_inquries)

# Percent of customer with changed credit limit
Percent_of_customer_with_changed_credit_limits =(data["Changed_Credit_Limit"]>0).mean()*100
print("Percent of customer with hanged credit limits:",Percent_of_customer_with_changed_credit_limits)

# Distribution of credit score(Good,Standard,Poor)
credit_score_counts = data["Credit_Score"].value_counts()
print("credit score counts:",credit_score_counts)
```

Behavioural Insights

```python
# Distribution of payment Behaviour
Distribution_of_payment_Behaviour = data["Payment_Behaviour"].value_counts()
print("Distribution of payment Behaviour: ",Distribution_of_payment_Behaviour)

# Percentage of customers who pay minimum amount only
Percent_minimumum_amount_only = (data["Payment_of_Min_Amount"] == "Yes").mean() * 100
print("Percent_minimumum_amount_only:", Percent_minimumum_amount_only)
```

 # Data Visualisation

Chart-1 Number of customers by Occoupation- Bar Chart

```python
Occoupation_wise_no_of_customers = data["Occupation"].value_counts()
print("Occoupation_wise_no_of_customers:",Occoupation_wise_no_of_customers)
```

```python
# Plot bar chart
plt.figure()
Occoupation_wise_no_of_customers.plot(kind='bar')
plt.xlabel('Occupation')
plt.ylabel('Number of Customers')
plt.title('Occupation-wise Number of Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 3))

ax.bar(
    x=Occoupation_wise_no_of_customers.index.astype(str),
    height=Occoupation_wise_no_of_customers.values,
    color='violet'
)

ax.set_xlabel('Occupation Name')
ax.set_ylabel('No. of Customers')
ax.set_title('Occupation wise No. of Customers')

ax.spines[['top', 'right', 'left']].set_visible(False)
plt.xticks(rotation=45)

plt.show()
```
