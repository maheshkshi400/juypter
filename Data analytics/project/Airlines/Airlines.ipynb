Q1 .what are the airlines inthe dataset,accompanied by their frequencies?

```python
import matplotlib.pyplot as plt

# Unique airline names
print(df['airline'].unique())

# Value counts
airline_counts = df['airline'].value_counts()
print(airline_counts)

# Bar chart
plt.subplot(1,2,1)
plt.bar(airline_counts.index, airline_counts.values)
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.title('Flight Count by Airline')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.subplot()
```

Q2.show bar grahs representing the Departure time and arrival time of the flights.

```python
import matplotlib.pyplot as plt

# Get value counts
dep_counts = df['departure_time'].value_counts()

# Plot bar graph
plt.subplot(1,2,1)
plt.figure(figsize=(7,5))
plt.bar(dep_counts.index, dep_counts.values)
plt.xlabel('Departure Time')
plt.ylabel('Count')
plt.title('Departure Time Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.subplot(1,2,2)
```

```python
import matplotlib.pyplot as plt

# Value counts of arrival_time
dep_counts = df['arrival_time'].value_counts()

# Bar graph
plt.figure(figsize=(7,5))
plt.bar(dep_counts.index, dep_counts.values)
plt.xlabel('Arrival Time')
plt.ylabel('Count')
plt.title('Arrival Time Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
