pivot & melting

```python
import  pandas as pd 
dict = {"keys":["k1","k2","k1","k2"],
        "names":["john","ben","David","peter"],
        "houses":["red","green","blue","yellow"]}
df= pd.DataFrame(dict)
print(df)

print(df.pivot(index="keys",columns="names",values="houses"))
```

```python
import pandas as pd 

data_dict = {
    "keys": ["k1", "k2", "k1", "k2"],
    "names": ["john", "ben", "David", "peter"],
    "houses": ["red", "green", "blue", "yellow"],
    "grades":["3rd","8th","9th","8th"]
}

df = pd.DataFrame(data_dict)
print(df)

# Using pivot_table to handle duplicates
pivot_df = df.pivot_table(index="keys", columns="names", values=["houses","grades"], aggfunc='first')
print(pivot_df)
```

```python
data_dict = {
    "names": ["john", "ben", "David", "peter"],
    "houses": ["red", "green", "blue", "yellow"],
    "grades":["3rd","8th","9th","8th"]
}

df = pd.DataFrame(data_dict)
print(df)


print(pd.melt(df,id_vars=["names"],value_vars=["houses"]))
```
