```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {"Days":[1,2,3,4,5],
        "NOP":[45,56,34,25,70]}
df= pd.DataFrame(data)
print(df)

sns.lineplot(data=data,x="Days",y="NOP")
plt.show()
```
