```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
data = pd.read_excel(r"C:\Users\User\juypter\Data analytics\expense3.xlsx")
df1 = pd.DataFrame(data)

# Print the DataFrame to check its contents (optional)
#print(df1)

# Group by 'Category' and sum the 'Amount'
group = df1.groupby("Category").agg({"Amount": "sum"})
print(group)

# Step plot
plt.figure(figsize=(10, 6))  # Set the figure size for the step plot
plt.step(group.index, group["Amount"], where="mid", marker="o", linestyle='-', color='b', label='Total Amount by Category')
plt.xlabel("Category")
plt.ylabel("Total Amount")
plt.title("Step Plot of Total Amount by Category")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add a grid for better readability
plt.legend()  # Show legend
plt.tight_layout()  # Adjust layout to make room for labels
plt.show()
```
