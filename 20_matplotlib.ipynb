```python
import matplotlib.pyplot as plt
import numpy as np

# types of plot-bar

x=[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
h=[9.8,7,12,3.9,7,6.5,8,9.5,11.5,3.2]               # height of the bars

c=["red","green","orange"]
plt.bar(x,h,width=5,align="center",color=c,linewidth=0.5,edgecolor="black")


plt.xlabel("year")
plt.ylabel("unemployment_rate")
plt.title("year vs unemployment_rate")

plt.show()
```
