```python
import matplotlib.pyplot as plt
import numpy as np

# types of plot-horizontal bar graph

y=[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
w=[9.8,7,12,3.9,7,6.5,8,9.5,11.5,3.2]               # height of the bars

c=["red","green","orange"]
plt.barh(y,w,5,align="center",color=c,linewidth=0.5,edgecolor="black")


plt.xlabel("year")
plt.ylabel("unemployment_rate")
plt.title("year vs unemployment_rate")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# types of plot-horizontal bar graph

x=["Russia","china","argentina","chillie","pakistan","Bharat"]

  
gold=[10,40,70,90,120,40]
silver=[108,200,205,107,302,204]
bronze=[110,107,103,127,105,310]

b_bronze=list(np.add(gold,silver))



plt.bar(x,gold,0.5,label="gold",color="red",edgecolor="black")
plt.bar(x,silver,0.5,bottom=gold,label="silver",color="yellow",edgecolor="black")
plt.bar(x,bronze,0.5,bottom=b_bronze,label="bronze",color="green",edgecolor="black")



plt.xlabel("countries")
plt.ylabel("medals")
plt.title("countries vs medals")
plt.legend()

plt.show()
```
